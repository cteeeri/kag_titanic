```python
# combine both train and test df to find mean Age and
# use the 4 features to better, hopefully, guess the age 
from decimal import Decimal

meanAge = int(Decimal(traindf['Age'].mean()))
meanAge
# Concatenate testdf to traindf to get better estimation of Age

withAge1 = traindf.dropna()
withAge2 = testdf.dropna()

withAge1and2 = pd.concat([withAge1, withAge2])

groupbyage = withAge1and2.groupby(['Pclass','SibSp','Parch', 'Sex'],as_index=False)['Age'].mean()
# groupbyage.isnull().sum()
groupbyage.info()
groupbyage
# b = groupbyage[(groupbyage['Pclass']== 1) & (groupbyage['SibSp']== 1) & (groupbyage['Parch']== 0) & (groupbyage['Sex']== 0)]['Age']
# b
```

```python
from decimal import Decimal

def guessAge(pclass, sibsp, parch, sex):
#     print("psps {} {} {} {}".format(pclass, sibsp, parch, sex))
    r = groupbyage[(groupbyage['Pclass']== pclass) & (groupbyage['SibSp']== sibsp) & (groupbyage['Parch']== parch) & (groupbyage['Sex']== sex)]['Age']
    if r.any():
        avg = r.values[0]
        return int(round(Decimal(avg),2)) 
    # could not find mean age for the specific combination of the 4 parameters
    return meanAge
```

```python
# traindf['Age'] = traindf.apply(lambda r: guessAge(r['Pclass'], r['SibSp'], r['Parch'], r['Sex']), axis=1)
def fillAge(r):
    age = r['Age']
    if pd.isnull(age):
        ga =  guessAge(r['Pclass'], r['SibSp'], r['Parch'], r['Sex'])
        return ga
    else:
        return age
    

# traindf['Age'] = traindf.apply(lambda r: guessAge(r['Pclass'], r['SibSp'], r['Parch'], r['Sex']) if r['Age'] == np.nan else pass, axis=1)


traindf['Age'] = traindf.apply(fillAge, axis=1)
```

```python
# name_df = pd.DataFrame(X_train[X_train['Age'].isnull()]['Name'])
# name_df = pd.DataFrame(X_train['Name'])
# name_df.info()
# name_array = name_df['Name'].values
X_train_good['Title'] = [np.nan]*len(X_train_good)
# X_train_good.info()
name_regex = r'.*(Mrs)|(Mr)|(Dr)|(Miss)|(Master)|(Col)|(Rev)|(Major)|(Ms)|(Capt)|(Countess)|(Mme).*'
import re
# for n in name_array:
#     r = re.search(name_regex, n, re.I)
#     if r: 
#         if r.group(1): 
#     else:
#         print(n)

# def copy_2title():


def repl_w_title(a,regex=name_regex):
    import re
    r = re.search(regex, a, re.I)
    if r:
        if r.group(1): return "Mrs"
        if r.group(2): return "Mr"
        if r.group(3): return "Dr"
        if r.group(4): return "Miss"
        if r.group(5): return "Master"
        if r.group(6): return "Col"
        if r.group(7): return "Rev"
        if r.group(8): return "Major"
        if r.group(9): return "Ms"
        if r.group(10): return "Capt"
        if r.group(11): return "Countess"
        if r.group(12): return "Mme"
    else:
        return a

# X_train_good[X_train_good['Name'].str.match(name_regex)]['Name'] #.apply(copy_2title)
X_train_good['Title'] = X_train_good['Name'].apply(repl_w_title,regex=name_regex)

X_train_good.info()
```

```python
from sklearn.metrics import r2_score

def performance_metric(y_true, y_predict):
    """ Calculates and returns the performance score between 
        true and predicted values based on the metric chosen. """
    
    # TODO: Calculate the performance score between 'y_true' and 'y_predict'
    score = r2_score(y_true, y_predict)
    
    # Return the score
    return score
```

```python
from sklearn import preprocessing

# 1. INSTANTIATE
# encode labels with value between 0 and n_classes-1.
le = preprocessing.LabelEncoder()
agedf_train_X_2 = agedf_train_X.apply(le.fit_transform)
agedf_train_X_2.head(5)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import ShuffleSplit, GridSearchCV
from sklearn.metrics.scorer import make_scorer

DTregressor = DecisionTreeRegressor()
cvSets = ShuffleSplit(n_splits=4, test_size=.2, random_state=0)
params = {'max_depth': range(1,5)}
scoring = make_scorer(performance_metric)
gscv = GridSearchCV(estimator=DTregressor,cv=cvSets,param_grid=params,scoring=scoring)
gscv = gscv.fit(agedf_train_X_2, agedf_train_y)

gscv.best_estimator_




# from sklearn.cross_validation import train_test_split
# ageX_train, ageX_test, agey_train, agey_test = train_test_split(agedf_train_X_2, agedf_train_y, test_size = 0.2, random_state = 10)
# DTregressor.fit(ageX_train,agey_train)
# age_score = DTregressor.score(ageX_test,agey_test)
# age_score
```
